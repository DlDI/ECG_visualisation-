# ECG 

```python
import neurokit2 as nk
import pandas as pd
import streamlit as st
from bokeh.plotting import figure, show
```

## First step : load and ecg file 


```python
ecg = nk.ecg_simulate(310, sampling_rate=1000)
df = pd.DataFrame(ecg)
```

```python
print(df)
```

```python
def ecg_preprocessing(df: pd.DataFrame):
    # peak detection
    df.squeeze()
    (read_ecg, dict_info) = nk.ecg_process(df, sampling_rate=1000, method="neurokit")
    simulated_ecg = read_ecg["ECG_Raw"]

    peaks = read_ecg["ECG_R_Peaks"]
    peak_index = [i for i in range(len(peaks)) if peaks[i] != 0]
    x_axis = range(0, len(simulated_ecg))

    return simulated_ecg, x_axis, peak_index, peaks
```

```python
# default values
freq = 1000
window_split = 10
segmentation = 1
selected = ecg
start = 0
end = len(selected)
```

```python
y, x, peak_index ,peaks = ecg_preprocessing(selected)
```

```python
from visualisation import get_figure
```

```python
fig = get_figure(x, y, start, end, peaks, window_split, freq, segmentation, peak_index)
```
